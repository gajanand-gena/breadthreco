```python
df[['Semester', 'Year']] = df['session'].str.split(' ', expand=True)
# df[['code', 'name']] = df['course'].str.split('-', expand=True)
df
```

```python
df[['Course Code', 'Course Name']] = df['course'].str.split('-',n= 1, expand=True)
```

```python
df.drop(['course','session'],axis =1 ,inplace =True)
```

```python
df
```

```python
breadth = pd.read_json("/kaggle/input/breadth-course/breadth_course.json")
breadth= breadth.transpose()
breadth=breadth[['name','prereq','L-T-P','id']]
```

```python
def extract_prerequisites(prereq_list):
    if isinstance(prereq_list, str):
        return prereq_list.strip('[]').split(', ') 
    else:
        return prereq_list  
breadth['Prerequisites_List'] = breadth['prereq'].apply(extract_prerequisites)
```

```python
breadth.head()
```

```python
breadth.columns
```

```python
breadth['L-T-P'] = breadth['L-T-P'].str.split('-').apply(lambda x: sum(map(int, x)))
```

```python
breadth.drop(['id','prereq'],axis=1,inplace =True)
breadth
```

```python
breadth.reset_index(inplace=True)
breadth.rename(columns={'index': 'Course Code'}, inplace=True)
```

```python
breadth
```

```python
df
```

```python
common_courses= df.merge(breadth, on='Course Code', how='inner')
final_df=common_courses.drop(['Course Name'],axis=1)
```

```python
final_df
```

```python
final_df['Department Code'] = final_df['Course Code'].str.extract(r'([A-Za-z]+)')
final_df['Course Number'] = final_df['Course Code'].str.extract(r'(\d+)')
```

```python
final_df.drop(['Course Code'],axis=1,inplace=True)
```

```python
final_df.to_csv('almost_done_df')
```
